```python
#Write a function that returns products whose price is at or below the threshold.
affordable_products = []
inventory = [
    {'name':'Sugar','price':130,'quantity':5},
     {'name':'Cooking Oil','price':350,'quantity':20}
    ]
price_threshold = 200

def get_affordable_products(inventory, price_threshold):
    affordable_products = []
    for product in inventory:
        if product["price"] <= price_threshold:
            affordable_products.append(product)
    return affordable_products


print(get_affordable_products(inventory,200))
```
